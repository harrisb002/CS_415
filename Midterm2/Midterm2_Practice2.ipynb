algorithm to determine this.

  - HINT: Show how to modify Dijkstra's algorithm. The only change is in the update step.

**Solution:**
- Dijkstra's algorithm is used for finding the shortest path in a graph with non-negative edge weights, and it operates in O((|V| + |E|) log |V|) time using a priority queue.
 - In this case, the algorithm needs to be adapted to find the smallest maximum
   edge weight (representing the minimum fuel tank capacity) in the path from 's' to 't'.

- Initialize a separate array, maxEdge, to record the max edge weight (fuel
  requirement) for the path from 's' to each vertex. Init maxEdge for each vertex as infinity, except for the start vertex 's', which should be initialized as 0.

- Use a priority queue to manage the vertices, with the key being the current
  max edge weight for the path to that vertex.
  
- In the standard Dijkstra's algorithm, you update the distance to neighboring
  vertices if the new computed distance is smaller than the current distance.
  In the modified algorithm, for each neighboring vertex 'v' of the current vertex 'u', calc newMaxEdge = max(maxedge[u], weight(u, v)), where weight(u, v) is the weight of the edge connecting 'u' and 'v'. Update maxEdge[v] only if newMaxEdge is < current maxEdge[v].

- Extract the vertex with the min maxEdge value from the priority queue.
 Update the maxEdge values of its adjacent vertices using the modified relax step above. After updating, if the maxEdge value for a vertex is reduced, decrease its key in the priority queue.

- The algorithm ends when the priority queue is empty or when the
 destination vertex 't' is taken out from the priority queue.
 The value maxEdge[t] gives the min fuel tank capacity needed to travel from 's' to 't'.

- This results in O((|V| + |E|) log |V|), as in Dijkstra's algorithm. The
  priority queue operations (extract-min and decrease-key) contribute to the log |V| factor, and each edge is considered once for the relax step, adding to the edge factor.

# Problem 7 (a)
- If some character occurs with frequency more than 0.4, there is guaranteed to be a codeword of length 1.



**Solution:**
- Assume you have a character 'A' with a freq. > 0.4.

- Since Huffman encoding gives shorter codes to more frequent chars, 'A' will be among the chars considered for shorter codes.

- Consider the last step where the two least frequent nodes are combined. For 'A' not to get a single-character code, there must be at least two other chars with a combined freq. higher than 0.4.

- However, this is impossible because if 'A' has a freq. > 0.4, the max combined freq. of any other two chars can only be < 0.6.

- Therefore, in the last step 'A' will be one of the two nodes combined, ensuring it has a code of length 1. (1 bit to be clear)

We will prove the contrapositive of the original statement, that is, that if there is a codeword of length 1, there must be a node with frequency greater than 1/3.

# Some Notes:

**Topological sorting**
  - Topological sorting applies to a ***directed graph*** and ***produces a
     linear ordering of its vertices*** such that for every directed edge u,v from vertex u to vertex v, u comes before v in the ordering.
  - Only possible if the ***graph has no cycles***, i.e., it is a Directed  
     Acyclic Graph (DAG). If there's a cycle, there's no linear order that can satisfy all the dependency requirements.
  - ***Commonly done using DFS.***
  - ***Not necessarily unique.*** There may be multiple valid orderings for a given graph.
